# Section 1: Data Preprocessing on Dataset


1. Given a dataset diabetes.csv, write a Python script to load and display the dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('diabetes.csv')
df.head()
```

2. Rename the columns accordingly: ‘Pregnancies’ to ‘Pregnant’, ‘BloodPressure’ to
‘BP’, ‘SkinThickness’ to ‘Skin’ and ‘DiabetesPedigreeFunction’ to ‘Pedigree’ and
display again.

```python
df.rename(columns={'Pregnancies':'Pregnant','BloodPressure':'BP','SkinThickness':'Skin','DiabetesPedigreeFunction':'Pedigree'},inplace=True)
df.head()
```

3. Use the describe() function to print the statistical summary of the data in the
dataframe

```python
df.describe()
```

4. Consider the ‘Pregnant’, ‘BP’, ‘Insulin’, ‘BMI’, ‘Pedigree’ and ‘Age’ to be the feature
columns and split the dataset into 80% train and 20% test data

```python
from sklearn.model_selection import train_test_split
x=df[['Pregnant','BP','Insulin','BMI','Pedigree','Age']]
y=df['Outcome']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

5. Create a scatterplot showing the relation between ‘BMI’ and ‘Age’ on the training data
with ‘Outcome’ as hue

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Age',y='BMI',hue='Outcome',data=pd.concat([x_train,y_train],axis=1))
plt.title('Scatterplot of BMI vs Age')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.show()
```

6. Perform Standardization using StandardScaler() function.

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
print(x_train.shape)
print(x_test.shape)
```

# Section 2: Logistic Regression Model

7. Train a logistic regression model using an inbuilt LogisticRegression() function on
train set.

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(x_train,y_train)
y_pred=logreg.predict(x_test)
```

8. Calculate the confusion matrix using inbuilt function confusion_matrix() and display
it using heatmap.

```python
from sklearn.metrics import confusion_matrix
cnf_matrix=confusion_matrix(y_test,y_pred)
cnf_matrix


class_names=[0,1]
sns.heatmap(pd.DataFrame(cnf_matrix),annot=True,cmap='YlGnBu',fmt='g')
plt.title('Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

 9. Calculate the accuracy and f1-score of the model using accuracy_score and f1_score
inbuilt methods respectively.

```python
from sklearn.metrics import accuracy_score,f1_score
print('Accuracy Score: ',accuracy_score(y_test,y_pred))
print('F1 Score: ',f1_score(y_test,y_pred))
```

10. Print the classification report using in classification_report() function with the target
names ‘with diabetes’ and ‘without diabetes’.

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred,target_names=['with diabetes','without diabetes']))
```
